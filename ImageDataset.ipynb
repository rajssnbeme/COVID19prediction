```python
import numpy as np
import pandas as pd
import cv2
import os
```

```python
data = pd.read_csv('/gdrive/MyDrive/Risk Prediction/Dataset/Mortality_incidence_sociodemographic_and_clinical_data_in_COVID19_patients.csv')
```

```python
data.head()
```

```python
data=data.drop('Age',axis=1)
data=data.drop('MAP',axis=1)
data=data.drop('PVD',axis=1)
data.head()
```

```python
dead_records=data[data['Death']==1]
dead_records.head()
```

```python
dead_records.shape
```

```python
dead_records_img=dead_records.drop(['Death'],axis = 1)
```

```python
dead_records_img.shape
```

```python
recovered_records=data[data['Death']==0]
recovered_records.head()
```

```python
recovered_records.shape
```

```python
recovered_records_img=recovered_records.drop(['Death'],axis = 1)
```

```python
recovered_records_img.shape
```

```python
def convert2image(row):
    img = row.values
    img = img.reshape(9,9)  # dimensions of the image
    image = np.zeros((9,9,1))  # empty matrix
    image[:,:,0] = img
    return image.astype(np.uint8)
```

```python
count1 = 0  # initialize counter
for i in range(1, dead_records_img.shape[0]):  #data.shape[0] gives no. of rows
  face = dead_records_img.iloc[i]  # remove one row from the data
  img = convert2image(face)  # send this row of to the function 
  # cv2.imshow("image", img) 
  # cv2.waitKey(0)  # closes the image window when you press a key   
  count1+=1  # counter to save the images with different name 
  if count1<=918:
    cv2.imwrite('/gdrive/MyDrive/Risk Prediction/Dataset/Image dataset/Train/Dead/'+ str(count1) +'.jpg',img) # path of location
  else:
    cv2.imwrite('/gdrive/MyDrive/Risk Prediction/Dataset/Image dataset/Test/Dead/'+ str(count1) +'.jpg',img)
```

```python
count2 = 0  # initialize counter
for i in range(1, recovered_records_img.shape[0]):  #data.shape[0] gives no. of rows
  face = recovered_records_img.iloc[i]  # remove one row from the data
  img = convert2image(face)  # send this row of to the function 
  # cv2.imshow("image", img) 
  # cv2.waitKey(0)  # closes the image window when you press a key   
  count2+=1  # counter to save the images with different name         
  if count2<=2850:
    cv2.imwrite('/gdrive/MyDrive/Risk Prediction/Dataset/Image dataset/Train/Recovered/'+ str(count2) +'.jpg',img) # path of location
  else:
    cv2.imwrite('/gdrive/MyDrive/Risk Prediction/Dataset/Image dataset/Test/Recovered/'+ str(count2) +'.jpg',img)
```
